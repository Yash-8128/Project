```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("/content/weather_classification_data.xlsx")
df
```

```python
# Basic Info
print("Shape of the Data:\n",df.shape)
print("\nColumn Name:\n",df.columns)
print("\nData Types:\n",df.dtypes)
```

```python
# Missing Values
print("\nMissing Values:\n",df.isnull().sum())
```

```python
# Duplicate Rows
print("\nDuplicate Rows:\n",df.duplicated().sum())
```

```python
# Descriptive Statisticts
print("\nDescriptive Statistics:\n",df.describe())
```

```python
# Value Counts (Categorical Columns)
print("\nSeason Value Counts:\n",df['Season'].value_counts())
print("\nLocation Value Counts:\n",df['Location'].value_counts())
print("\nWeather Value Counts:\n",df['Weather Type'].value_counts())
```

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
plt.imshow(corr_matrix, interpolation='nearest')
plt.title("Correlation Heatmap")
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.show()
```

```python
# Histrogram of Numerical Columns
df.hist(figsize=(12,10))
plt.suptitle("Distribution of numerical columns")
plt.show()
```

```python
# Season wise Tempature Boxplot
plt.figure(figsize=(8,5))
df.boxplot(column="Temperature",by="Season")
plt.suptitle("")
plt.show()
```

```python
# Weather Type Count Plot
df['Weather Type'].value_counts().plot(kind='bar', figsize=(8,5))
plt.title("Weather Type Frequency")
plt.xlabel("Weather Type")
plt.ylabel("Count")
plt.show
```
